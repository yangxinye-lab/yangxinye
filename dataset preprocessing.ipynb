```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from imageio import imread, imwrite
from PIL import Image
```

```python
#Sequentially read different image category folders under the image path
image_file='images/'
path_list=os.listdir(image_file)
path_list.remove('.DS_Store')
path_list.sort(key=lambda x:int(x.split('.')[0]))
print(path_list)
```

```python
#get all images under each image category and check the image size and number of joints
def each_class_pic(path,path_list):
    for i in path_list:
        cur_bird_list=os.listdir(os.path.join(path,i))
        for img in cur_bird_list[7:8]:
            cur_bird_img=Image.open(os.path.join(os.path.join(path,i),img))
            #resize picture 
            if cur_bird_img.size[0]!=256 and cur_bird_img.size[1]!=256:
                cur_bird_img=convertjpg(cur_bird_img)
                print('cur_img_name:',img,'img size：',cur_bird_img.size,'img mode：',cur_bird_img.mode)
                #convert pic mode
                if(cur_bird_img.mode!='RGB'):
                    cur_bird_img=cur_bird_img.convert('RGB')
                    print('cur_img_name:',img,'img size：',cur_bird_img.size,'img mode：',cur_bird_img.mode)
                    #save picture to a new path
                    if not os.path.exists(os.path.join('new_images',i)):
                        os.makedirs(os.path.join('new_images',i))
                    cur_bird_img.save(os.path.join(os.path.join('new_images',i),img))

#Visualization picture
def plt_img(img):
    img=np.array(img)
    plt.imshow(img)
    plt.show()
    
#Convert the picture to a uniform size
def convertjpg(img,width=256,height=256):
    new_img=img.resize((width,height),Image.BILINEAR) 
    return new_img

    
each_class_pic(image_file,path_list)
        
```

```python
#Add random noise (actually, randomly change the gray value of pixels on the image to 255, which is white)
def random_noise(image,noise_num):
    img = cv2.imread(image)
    img_noise = img
    rows, cols, chn = img_noise.shape
    for i in range(noise_num):
        x = np.random.randint(0, rows)
        y = np.random.randint(0, cols)
        img_noise[x, y, :] = 255
    return img_noise

for i in path_list[59:60]:
    cur_bird_list=os.listdir(os.path.join(image_file,i))
    for img in cur_bird_list[10:11]:
        img_path=os.path.join(os.path.join(image_file,i),img)
        plt_img(Image.open(img_path))
        img_noise=random_noise(img_path,3000)
        plt_img(img_noise)
```

```python
def gamma_transform(image):
    img = cv2.imread(image, 0)
    fi = img / 255.0
    gamma = 0.4
    out_img = np.power(fi, gamma)
    return out_img

for i in path_list[59:60]:
    cur_bird_list=os.listdir(os.path.join(image_file,i))
    for img in cur_bird_list[20:21]:
        img_path=os.path.join(os.path.join(image_file,i),img)
        plt_img(Image.open(img_path))
        gamma_img=gamma_transform(img_path)
        plt_img(gamma_img)
```

```python
def pic_transpose(image):
    img = Image.open(image)
#     transpose_img = img.transpose(Image.ROTATE_90)
    transpose_img = img.transpose(Image.ROTATE_180)
    #transpose_img = img.transpose(Image.ROTATE_270)
    return transpose_img

for i in path_list[59:60]:
    cur_bird_list=os.listdir(os.path.join(image_file,i))
    for img in cur_bird_list[33:34]:
        img_path=os.path.join(os.path.join(image_file,i),img)
        plt_img(Image.open(img_path))
        transpose_img=pic_transpose(img_path)
        plt_img(transpose_img)
```

```python
import math
def get_imghist(img):
    if len(img.shape) == 3:
        img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    hist = cv2.calcHist([img],[0],None,[256],[0,255])
    return hist

def cal_equalhist(img):
    img=cv2.imread(img)
    if len(img.shape) == 3:
         img= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    h,w = img.shape[:2]
    grathist = get_imghist(img)
 
    zerosumMoment = np.zeros([256],np.uint32)
    for p in range(256):
        if p ==0:
            zerosumMoment[p] = grathist[0]
        else:
            zerosumMoment[p] = zerosumMoment[p-1] +grathist[p]
 
    output_q = np.zeros([256],np.uint8)
    cofficient = 256.0/(h*w)
    for p in range(256):
        q = cofficient *float(zerosumMoment[p]) - 1
        if q >= 0:
            output_q[p] = math.floor(q)
        else:
            output_q[p] = 0
 
    equalhistimage = np.zeros(img.shape,np.uint8)
    for i in range(h):
        for j in range(w):
            equalhistimage[i][j] = output_q[img[i][j]]
    return equalhistimage

```

```python
for i in path_list[59:60]:
    cur_bird_list=os.listdir(os.path.join(image_file,i))
    for img in cur_bird_list[12:13]:
        img_path=os.path.join(os.path.join(image_file,i),img)
        plt_img(Image.open(img_path))
        transpose_img=cal_equalhist(img_path)
        plt_img(transpose_img)
```
